7. __What is the difference between instance, static and class methods in python?__  
    - instance methods are methods that are bound to an instance of a class. They can access and modify the instance's attributes and state, and can be called on any instance of the class. Instance methods are defined using the self parameter, which refers to the instance that the method is called on.  
    - Class methods, on the other hand, are methods that are bound to the class itself, rather than to an instance of the class. They can access and modify class-level attributes and state, and can be called on the class itself, rather than on an instance of the class. Class methods are defined using the @classmethod decorator, and take a <b>cls</b> parameter, which refers to the class that the method is called on.
    With staticmethods, neither self (the object instance) nor cls (the class) is implicitly passed as the first argument. They behave like plain functions except that you can call them from an instance or the class.  
    A class method can access or modify the class state while a static method can’t access or modify it.
    We generally use the class method to create factory methods. Factory methods return class objects ( similar to a constructor ) for different use cases.  
    - Static methods : use @staticmethod decorator, are not related to a specific instance, and are self-contained, We generally use static methods to create utility functions.

    ```python
    class MyClass:
        count = 0

        def __init__(self):
            MyClass.count += 1

        @classmethod
        def get_count(cls):
            return cls.count
        
        @staticmethod
        def static_func(x):
            return x

    # create some instances of MyClass
    obj1 = MyClass()
    obj2 = MyClass()
    obj3 = MyClass()

    # call the class method on the class itself
    result = MyClass.get_count()

    print(result)  # Output: 3
    print(obj1.static_func(1))  # Output: 1
    print(MyClass.static_func(1))  # Output: 1
    #############################################################################
    ```

    ```python
    class CoffeeShop:
        specialty = 'espresso'
        
        def __init__(self, coffee_price):
            self.coffee_price = coffee_price
        
        # instance method
        def make_coffee(self):
            print(f'Making {self.specialty} for ${self.coffee_price}')
        
        # static method    
        @staticmethod
        def check_weather():
            print('Its sunny')
        # class method
        @classmethod
        def change_specialty(cls, specialty):
            cls.specialty = specialty
            print(f'Specialty changed to {specialty}')
    ```
    CoffeeShop class has an attribute, specialty, set to 'espresso' by default. Each instance of CoffeeShop is initialized with an attribute coffee_price . It also has 3 methods, an instance method, a static method and a class method.

    Let’s initialize an instance of the coffee shop with a coffee_price of 5. Then call the instance method make_coffee.
    ```python
    coffee_shop = CoffeeShop('5')
    coffee_shop.make_coffee()
    #=> Making espresso for $5
    ```
    Now call the static method. Static methods can’t modify class or instance state so they’re normally used for utility functions, for example, adding 2 numbers. We used ours to check the weather.Its sunny. Great!
    ```python
    coffee_shop.check_weather()
    #=> Its sunny
    ```
    Now let’s use the class method to modify the coffee shop’s specialty and then make_coffee.
    ```python
    coffee_shop.change_specialty('drip coffee')
    #=> Specialty changed to drip coffee
    coffee_shop.make_coffee()
    #=> Making drip coffee for $5
    ```
    Note how make_coffee used to make espresso but now makes drip coffee!

14. __What is the difference between a shallow and a deep copy?__  

We’ll walk through 3 scenarios.

1. Reference the original object. This points a new name, li2, to the same place in memory to which li1 points. So any change we make to li1 also occurs to li2.
    ```python
    li1 = [['a'],['b'],['c']]
    li2 = li1
    li1.append(['d'])
    print(li2)
    #=> [['a'], ['b'], ['c'], ['d']]
    ```
2. Create a shallow copy of the original. We can do this with the list() constructor, or the more pythonic mylist.copy() 
A shallow copy creates a new object, but fills it with references to the original. So adding a new object to the original collection, li3, doesn’t propagate to li4, but modifying one of the objects in li3 will propagate to li4.
```python
    li3 = [['a'],['b'],['c']]
    li4 = list(li3)
    li3.append([4])
    print(li4)
    #=> [['a'], ['b'], ['c']]
    li3[0][0] = ['X']
    print(li4)
    #=> [[['X']], ['b'], ['c']]
```
3. Create a deep copy. This is done with copy.deepcopy(). The 2 objects are now completely independent and changes to either have no affect on the other.
    ```python
    import copy
    li5 = [['a'],['b'],['c']]
    li6 = copy.deepcopy(li5)
    li5.append([4])
    li5[0][0] = ['X']
    print(li6)
    #=> [['a'], ['b'], ['c']]
    ```

54. In Python, a generator is a special type of iterator that generates values on-the-fly using the yield keyword. Unlike regular functions that return a value and then exit, generators can pause and resume execution, allowing them to produce a sequence of values over time.  
we can use next() to acces values of a generator  
Here's an example of a simple generator function that generates the first n even numbers:  

    ```python
    def even_numbers(n):
        i = 0
        while i < n:
            yield i * 2
            i += 1


    #// create a generator that generates the first 5 even numbers
    gen = even_numbers(5)

    #// iterate over the values generated by the generator
    for num in gen:
        print(num)
    #// Output: 0, 2, 4, 6, 8

    next(gen)
    #//output: 0
    ```

59. Encapsulation is a fundamental concept in object-oriented programming that refers to the practice of hiding the internal details of an object from the outside world and restricting access to its internal state. In Python, encapsulation can be achieved through the use of private and protected attributes and methods.

60. private attributes and methods are denoted by a double underscore prefix (__) in their names. They cannot be accessed directly from outside the class, and attempting to do so will result in an AttributeError. Private attributes and methods can only be accessed from within the class itself.  
Protected attributes and methods are denoted by a single underscore prefix (_) in their names. They can be accessed from outside the class. can be accessed from within the class itself and from its subclasses.  

    ```python
    class MyClass:
        def __init__(self, x, y):
            self.__x = x       #pivate attr   #no access from instances
            self._y = y        #protected attr

        def __my_private_method(self):
            print("This is a private method")   #no access from instances

        def _my_protected_method(self):
            print("This is a protected method")
        
        def my_public_method(self):
            self._my_protected_method()
            print("This is a public method")

    class MySubclass(MyClass):
        def my_subclass_method(self):
            self._my_protected_method()
    ```
